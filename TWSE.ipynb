```python
import pandas as pd
import requests
import xlwings as xw
from time import sleep
```

```python
url = 'https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=20210601&type='
```

```python
cleaned_data = []
```

```python
for i in range(1,10):
    res = requests.get(url+'0'+str(i))
    data = res.text
    for da in data.split('\n'):
        if len(da.split('","')) == 16 and da.split('","')[0][0] != '=':
            cleaned_data.append([ele.replace('",\r','').replace('"','') 
                             for ele in da.split('","')])
    sleep(5)
```
